```python
from tabulate import tabulate

# Total de pessoas na pesquisa
total_pessoas_pesquisa = df['ID'].nunique()

# Filtrando apenas as pessoas que estavam na rua durante a entrevista (não em casa de acolhimento)
total_na_rua = df[df['Rua_Acolhimento'] == 'Rua']['ID'].nunique()

porcentagem_na_rua = ((total_na_rua / total_pessoas_pesquisa) * 100)
porcentagem_na_rua = round(porcentagem_na_rua, 1)
cards = [
    ["Total de pessoas na pesquisa", total_pessoas_pesquisa],
    ["Total de pessoas na rua", total_na_rua],
    ["Porcentagem de pessoas na rua (%)", porcentagem_na_rua]
]

# Imprimindo os resultados como cards
print("Resultados:")
print(tabulate(cards, tablefmt="grid"))
```

```python
labels = ['Total na rua', 'Total na pesquisa']
valores = [total_na_rua, total_pessoas_pesquisa]

plt.bar(labels, valores, color=['blue', 'orange'])
plt.ylabel('Número de Pessoas')
plt.title('Comparação entre Total de Pessoas na Rua e na Pesquisa')

for i, valor in enumerate(valores):
    plt.text(i, valor + 20, f'{valor} ({porcentagem_na_rua}%)', ha='center', va='bottom')

plt.show()
```

```python

df_feminino = df[df['Sexo'] == 'Feminino']

total_mulheres = df_feminino.shape[0]

total_gravidas = df_feminino[df_feminino['Gravidez'] == 'Sim'].shape[0]

porcentagem_gravidas = (total_gravidas / total_mulheres) * 100

total_diabetes = df_feminino[(df_feminino['Gravidez'] == 'Sim') & (df_feminino['Problema_Saude_Diabetes'] == 'Sim')].shape[0]
total_hiv = df_feminino[(df_feminino['Gravidez'] == 'Sim') & (df_feminino['Problema_Saude_HIV_AIDS'] == 'Sim')].shape[0]
total_pressao_alta = df_feminino[(df_feminino['Gravidez'] == 'Sim') & (df_feminino['Problema_Saude_Pressao_alta'] == 'Sim')].shape[0]
total_acompanhamento_prenatal = df_feminino[(df_feminino['Gravidez'] == 'Sim') & (df_feminino['Acompanhamento_pre_natal'] == 'Sim')].shape[0]

porcentagem_acompanhamento_prenatal = (total_acompanhamento_prenatal / total_gravidas) * 100 if total_gravidas > 0 else 0

print(f'Total de mulheres: {total_mulheres}')
print(f'Total de mulheres grávidas: {total_gravidas} ({porcentagem_gravidas:.2f}%)')
print(f'Total de grávidas com diabetes: {total_diabetes}')
print(f'Total de grávidas com HIV/AIDS: {total_hiv}')
print(f'Total de grávidas com pressão alta: {total_pressao_alta}')
print(f'Total de grávidas com acompanhamento pré-natal: {total_acompanhamento_prenatal} ({porcentagem_acompanhamento_prenatal:.2f}%)')
```

* Todas as grávidas terem recebido acompanhamento pré-natal sugere um acesso relativamente bom aos serviços de saúde ou a eficácia de programas de apoio social e saúde específicos para essa população vulnerável.

* Condições de Saúde das Mulheres Grávidas

      - Embora os números sejam relativamente baixos, são preocupantes as condições de saúde enfrentadas durante a gravidez.

```python
deficiencias = ['Deficiencia_Caminhar_ou_degraus', 'Deficiencia_Enxergar',
                'Deficiencia_Ouvir', 'Deficiencia_Mental', 'Deficiencia_Nao_possui']
porcentagem_deficiencia = []

for deficiencia in deficiencias:
    porcentagem = df_na_rua[deficiencia].value_counts(normalize=True).get('Sim', 0) * 100
    porcentagem_deficiencia.append(porcentagem)

# Preparando os dados para o gráfico
dados_deficiencia = pd.DataFrame({
    'Deficiência': deficiencias,
    'Porcentagem': porcentagem_deficiencia
})

dados_deficiencia = dados_deficiencia.sort_values(by='Porcentagem', ascending=False)

plt.figure(figsize=(8, 4))
sns.set_style("whitegrid")  # Estilo de fundo

bars = sns.barplot(x='Porcentagem', y='Deficiência', data=dados_deficiencia, palette='viridis')

for bar, porcentagem in zip(bars.patches, dados_deficiencia['Porcentagem']):
    plt.text(bar.get_width() + 1, bar.get_y() + bar.get_height()/2, f'{porcentagem:.1f}%',
             va='center', ha='left', fontsize=10)

# Configurações adicionais
plt.title('Porcentagem de Pessoas com Deficiência na Rua', fontsize=12)
plt.xlabel('Porcentagem (%)', fontsize=10)
plt.ylabel('Tipo de Deficiência', fontsize=10)
plt.xlim(0, 100)  # Limitando o eixo x de 0 a 100

plt.tight_layout()
plt.show()
```

* 35% da população não possui nenhum tipo de deficiência;

* 16% das pessoas em situação de rua têm deficiência visual ou auditiva.

      - Importância de serviços de saúde e suporte adaptados para pessoas com deficiência sensorial.
      
      - Acessibilidade a serviços de saúde e orientações para melhorar a qualidade de vida e a inclusão desses indivíduos.

### Análise das pessoas em situação de rua - usuários de drogas

```python
df_filtrado = df_na_rua[df_na_rua['Faz_uso_drogas'] != 'Não se aplica']

uso_drogas_sexo = df_filtrado.groupby('Sexo')['Faz_uso_drogas'].value_counts(normalize=True).mul(100).rename('Porcentagem').reset_index()

plt.figure(figsize=(8, 6))
sns.barplot(x='Sexo', y='Porcentagem', hue='Faz_uso_drogas', data=uso_drogas_sexo, palette='Set2')
plt.title('Uso de Drogas - Sexo')
plt.ylabel('Porcentagem (%)')
plt.xlabel('Sexo')
plt.legend(title='Faz uso de drogas', loc='upper right')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

* Quase 100% dos indivíduos do sexo masculino e aproximadamente 85% das mulheres são usuários de drogas.

      - Problemas de saúde mental não tratados;
      - Falta de suporte social e econômico;
      - Estresse crônico.
      - Estratégia de enfrentamento para lidar com traumas passados ou violência de gênero.

```python
# Análise de uso de drogas por raça
uso_drogas_raca = df_filtrado.groupby('Cor_raca')['Faz_uso_drogas'].value_counts(normalize=True).mul(100).rename('Porcentagem').reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='Cor_raca', y='Porcentagem', hue='Faz_uso_drogas', data=uso_drogas_raca, palette='Set2')
plt.title('Uso de Drogas por Raça')
plt.ylabel('Porcentagem (%)')
plt.xlabel('Raça')
plt.legend(title='Faz uso de drogas', loc='upper right')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

* Independentemente da raça, sempre mais de 80% dos indivíduos em situação de rua são usuários de drogas.

      - Pobreza, falta de moradia ou falta de acesso a serviços de saúde mental não tem cor.

```python
# Análise do uso de drogas por faixa etária
uso_drogas_faixa_etaria = df_filtrado.groupby('Faixa_etaria')['Faz_uso_drogas'].value_counts(normalize=True).mul(100).rename('Porcentagem').reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(x='Faixa_etaria', y='Porcentagem', hue='Faz_uso_drogas', data=uso_drogas_faixa_etaria, palette='Set2')
plt.title('Uso de Drogas por Faixa Etária')
plt.ylabel('Porcentagem (%)')
plt.xlabel('Faixa Etária')
plt.legend(title='Faz uso de drogas', loc='upper right')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

* 100% das pessoas até 11 anos são usuárias;

* 90% das pessoas na faixa etária de 18 a 30 são usuárias.

```python
todas_drogas = [
    'Drogas_Tabaco',
    'Drogas_Alcool',
    'Drogas_Maconha_Haxixe',
    'Drogas_Crack_Similares',
    'Drogas_Cocaina'
]

df_filtrado = df_na_rua[df_na_rua['Faz_uso_drogas'] != 'Não se aplica']

uso_drogas = df_filtrado[todas_drogas].apply(lambda x: x.value_counts()['Sim'], axis=0)

top_3_drogas = uso_drogas.sort_values(ascending=True).head(3)

plt.figure(figsize=(8, 6))
top_3_drogas.plot(kind='barh', color=['#1f77b4', '#ff7f0e', '#2ca02c'])
plt.title('Três Drogas Mais Utilizadas')
plt.xlabel('Número de Pessoas')
plt.ylabel('Tipo de Droga')
plt.tight_layout()
plt.show()
```

A prevalência de maconha, cocaína e crack entre a população em situação de rua destaca a gravidade e a complexidade do uso de substâncias nessa comunidade.

* Maconha, acessível e socialmente mais aceita, junto com cocaína e crack, que possuem altos índices de dependência e consequências devastadoras para a saúde, exige uma abordagem integrada nas políticas públicas.

       - Implementar programas de tratamento e reabilitação, com ênfase na redução de danos e no suporte psicossocial;
       - Campanhas educativas contínuas para mitigar os riscos e consequências dessas drogas.
